```python
# Import the modules
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
# Convert the csv data into dataframe
data = pd.read_csv("seattle-weather.csv")
data.head()
```

```python
# Data Pre-processing
# Check for null values using isnull()
data.isnull().sum()
```

```python
# Transfer categorical data into numbers using label encoder.
def label_encoding(c):
    le = preprocessing.LabelEncoder()
    data[c] = le.fit_transform(data[c])
    data[c].unique
label_encoding("weather")
data
```

```python
cols = ["precipitation", "temp_max", "temp_min", "wind"]
```

```python
# Normalise the data
def normalise_data(df, cols):
    for x in cols:
        df[x] = df[x]/df[x].max()

normalise_data(data,cols)
data
```

```python
data = data.drop('date',axis=1)
data
```

```python
#Dividing the data
x =  data.drop("weather", axis = 1)
y = data['weather']
```

```python
# Splitting the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Machine leraning model
xg = XGBClassifier()
xg.fit(x_train,y_train)
```

```python
y_pre = xg.predict(x_test)
y_pre
```

```python
y_test
```

```python
# Checking the accuracy
print("Accuracy Score: ", accuracy_score(y_test, y_pre))
print("Clasification Report: ", classification_report(y_test, y_pre))
      
```

```python
# Tuning the parameters gamma, learning rate

grid = {'learning_rate':[0.1, 1, 0.01, 0.001], 'gamma':[0, 1, 10, 100]}
```

```python
model = GridSearchCV(xg, grid, cv= 10, verbose=2)
```

```python
model.fit(x_train, y_train)
```

```python
grid_predict = model.predict(x_test)
print(grid_predict)
```

```python
# Checking the accuracy again ||| Model Evaluation
print("Accuracy Score: ", accuracy_score(y_test, grid_predict))
print("Clasification Report: ", classification_report(y_test, grid_predict))
```
